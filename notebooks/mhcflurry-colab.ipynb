# Predict for specified peptides

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
IEDBBenchmark= pd.read_csv("/content/drive/MyDrive/totalresultofIEDBBenchmark.csv")
```

```python
grouped = IEDBBenchmark.groupby('Allele')['Peptide seq'].apply(list).reset_index()
```

```python
grouped
```

```python
templist = []
for i in range(0,grouped.shape[0]):
  Alleles=grouped[i:i+1]['Allele'].tolist()
  peptides=grouped[i:i+1]['Peptide seq'].tolist()[0]
  print(peptides)
  results1 = predictor.predict(peptides, Alleles)['affinity'].tolist()
  templist.append(results1)
```

```python
grouped['result']=templist
```

```python
final_result=grouped.explode(['Peptide seq', 'result']).reset_index(drop=True)
```
